Exercise 3.1 — FizzBuzz

Print the numbers 1 to 50. But: for multiples of 3 print "Fizz" instead, for multiples of 5 print "Buzz", and for multiples of both 3 and 5 print "FizzBuzz". This is the classic programming exercise — every developer has done it.


```python
for i in range(1,51):
    if i%15 == 0:
        print("FizzBuzz")
    elif i%5 == 0:
        print("Buzz")
    elif i%3 == 0:
        print("Fizz")
    else:
        print(i)
```

Exercise 3.2 — Order filter with comprehensions

Given a list of 10 order values (mix of numbers from 50 to 6000 — make them up), use list comprehensions to produce three new lists:

— All orders above £1000
— All orders with a 10% discount applied (all orders, not filtered)
— Orders classified as "small" (<500), "medium" (500–2000), or "large" (>2000) — one string label per order

Then print each list with a descriptive label.

```python
orderval = [85,92,108,352,5623,982,864,450,2980,1000]

above1000 = [o for o in orderval if o>1000]
discval = [o*0.9 for o in orderval]
classification = ["small" if o <500 else "medium" if o<2000 else "large" for o in orderval]

print(f"Orders above 1000: {above1000}")
print(f"Discounted order value: {discval}")
print(f"Order Classification: {classification}")
```

Exercise 3.3 — Number guessing game

Build a complete number guessing game that uses everything from Days 1, 2, and 3:

— Python picks a random number between 1 and 100 (use import random; random.randint(1,100))
— Ask the user to guess in a while loop
— Tell them "too high", "too low", or "correct!" with each guess
— Count the number of guesses
— When they win, print their guess count and rate it: 1–3 guesses = "Excellent", 4–6 = "Good", 7+ = "Keep practising"
— If they type "quit", exit the game early

```python
import random

num = random.randint(1,100)
n = 0
print({"Guess a number between 1 and 100. Type 'quit' to exit the game"})
while True:
    n +=1
    i = input("Please Enter Your Guess:")
    if i == "quit":
        n = -1
        break
    else:
        if(int(i) > num):
            print("Too High")
        elif(int(i) < num):
            print("Too Low")
        elif(int(i) == num):
            print("Correct!")
            break

if n == -1:
    print("Game Quit")
elif n <= 3:
    print(f"Guess Count: {n} - Excellent")
elif n <=6:
    print(f"Guess Count: {n} - Good")
else:
    print(f"Guess Count: {n} - Keep Practicing")

print(f"The number was - {num}")
```

Implementing a binary search strategy

```python
import random

num = random.randint(1,100)
print(f"The number to guess is: {num}")
guess = 50
minguess = 1
maxguess = 100
n = 0
while True:
    n += 1
    print(f"The guess is - {guess}")
    if guess > num:
        print("Too High")
        maxguess = guess
    elif guess < num:
        print("Too Low")
        minguess = guess
    else:
        print("Correct Guess")
        break

    print(f"Max Guess - {maxguess}")
    print(f"Min Guess - {minguess}")

    guess = round((maxguess + minguess)/2,0)

print(f"Total Number of guesses - {n}")
```
