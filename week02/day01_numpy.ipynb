Exercise W2.1 — Vectorised sales analysis

Create a NumPy array of 14 days of sales figures (make up values between 500 and 5000). Using only NumPy operations — no loops — compute and print:

— Total, mean, median, std deviation, min, max
— The index (day number) of the best and worst day
— All days that were above the mean
— Sales from week 1 (days 1–7) vs week 2 (days 8–14) total
— Percentage change: (week 2 − week 1) / week 1 × 100

```python
import numpy as np

np.random.seed(42)

sales = np.random.randint(500,5000, size = 14)
print(f"Sales: {sales}\n")

print(f"Total Sales: {np.sum(sales)}")
print(f"Mean: {np.mean(sales):.1f}")
print(f"Median: {np.median(sales):.1f}")
print(f"Std Dev: {np.std(sales):1f}")
print(f"Max Sales: {np.max(sales)}")
print(f"Min Sales: {np.min(sales)}")

print(f"Best day index: {np.argmax(sales)+1} ({sales[np.argmax(sales)]})")
print(f"Worst day index: {np.argmin(sales)+1} ({sales[np.argmin(sales)]})")


print(f"Sales greater than mean: {sales[sales > sales.mean()]}")

print(f"Total sales of week 1: {np.sum(sales[0:7])}")
print(f"Total sales of week 2: {np.sum(sales[7:])}")

week1sales = np.sum(sales[0:7])
week2sales = np.sum(sales[7:])
print(f"Percentage Change: {((week2sales - week1sales)/week1sales) * 100:.1f}%")
```

Exercise W2.2 — 2D matrix operations

Create a 6×4 NumPy array representing 6 months of sales for 4 products (use np.random.randint to generate it with seed=0). Then compute, using axis operations — no loops:

— Total revenue per product (across all months)
— Total revenue per month (across all products)
— Best performing product (highest total)
— Worst performing month (lowest total)
— The percentage share of each product's total out of the grand total

```python
import numpy as np

np.random.seed(0)
matrix   = np.random.randint(500, 5000, size=(6, 4))
products = ["Widget A", "Widget B", "Gadget X", "Connector"]
months   = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

prod_totals  = np.sum(matrix, axis=0)
month_totals = np.sum(matrix, axis=1)
grand_total  = np.sum(matrix)
shares       = prod_totals / grand_total * 100

print("=== Revenue by product ===")
for name, total, share in zip(products, prod_totals, shares):
    print(f"  {name:10} £{total:>7,}  ({share:.1f}%)")

print("\n=== Revenue by month ===")
for month, total in zip(months, month_totals):
    print(f"  {month} £{total:>7,}")

print(f"\nBest product:  {products[np.argmax(prod_totals)]}")
print(f"Worst month:   {months[np.argmin(month_totals)]}")
print(f"Grand total:   £{grand_total:,}")
```

Exercise W2.3 — Forecast accuracy calculator

Given two arrays — actual demand and forecast demand for 12 months — compute the following accuracy metrics using only NumPy vectorised operations:

— MAE (Mean Absolute Error)
— RMSE (Root Mean Square Error)
— MAPE (Mean Absolute Percentage Error) — skip any month where actual is 0
— BIAS (mean of signed errors, positive = over-forecast)
— Flag months where the absolute percentage error exceeds 20% as "high error months"
— Print a month-by-month table showing actual, forecast, and percentage error

```python
import numpy as np

month = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
np.random.seed(42)

actualDemand = np.random.randint(500,5000,12)

np.random.seed(0)

forecastDemand = np.random.randint(500,5000,12)

print(f"Actual Demand: {actualDemand}")
print(f"Forecasted Demand: {forecastDemand}")

error = (forecastDemand - actualDemand)

print(f"Errors: {error}")

mae = np.mean(np.abs(error))
print(f"Mean Absolute Error: {mae:.1f}")

mape = np.mean(np.abs(error)/actualDemand * 100)
print(f"Mean Absolute Percentage Error: {mape:.2f}")

rmse = np.sqrt(np.sum(error**2)/error.size)
print(f"Root mean squared error: {rmse:.1f}")

print(f"Bias: {np.mean(error)}")

errorflag = np.where(np.abs(error)/actualDemand * 100>20,"High Error Month","Low Error Month")

print(f"High Error Flag: {errorflag}")

print(f"Month - ActualDemand Forecast Demand PercentageError Flag")
for month,actual,forecast,error,flag in zip(month,actualDemand,forecastDemand,error,errorflag):
    print(f"{month} - {actual} {forecast} {error/actual * 100:.2f}% {flag}")
```

```python
#Alternate Solution for above exercise
import numpy as np

actual   = np.array([1200, 1350, 980, 1500, 1100, 1800,2100, 950,  1400, 1650, 1250, 1900])
forecast = np.array([1150, 1400, 1050, 1480, 1300, 1750,1950, 1100, 1380, 1700, 1100, 2050])

months    = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

errors    = forecast - actual
abs_errs  = np.abs(errors)
pct_errs  = abs_errs / actual * 100

mae  = np.mean(abs_errs)
rmse = np.sqrt(np.mean(errors**2))
mape = np.mean(pct_errs[actual != 0])
bias = np.mean(errors)

print("=== Forecast Accuracy Report ===")
print(f"MAE:  {mae:.1f} units")
print(f"RMSE: {rmse:.1f} units")
print(f"MAPE: {mape:.2f}%")
print(f"BIAS: {bias:+.1f} ({'over-forecast' if bias>0 else 'under-forecast'})")

print(f"\n{'Month':5} {'Actual':>7} {'Forecast':>9} {'Error%':>7} {'Flag'}")
print("-"*38)
for i, m in enumerate(months):
    flag = "HIGH" if pct_errs[i] > 20 else ""
    print(f"{m:5} {actual[i]:>7,} {forecast[i]:>9,} {pct_errs[i]:>6.1f}% {flag}")
```
