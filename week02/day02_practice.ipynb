```python
import pandas as pd
import numpy as np

# From a list — Pandas auto-creates integer index 0,1,2...
sales = pd.Series([1200, 850, 2100, 975, 1800])
print(sales)
# 0    1200
# 1     850
# 2    2100
# 3     975
# 4    1800
# dtype: int64

# With a custom index — much more useful
sales = pd.Series(
    [1200, 850, 2100, 975, 1800],
    index=["Mon", "Tue", "Wed", "Thu", "Fri"]
)
print(sales["Wed"])    # 2100 — access by label
print(sales.mean())    # 1385.0
print(sales * 1.1)     # vectorised — 10% increase
```

```python
import pandas as pd
import numpy as np

# Create a realistic supply chain dataset
np.random.seed(42)
n = 200

products   = ["Widget A", "Widget B", "Gadget X", "Connector", "Sensor Y"]
categories = {"Widget A":"Hardware", "Widget B":"Hardware","Gadget X":"Electronics", "Connector":"Hardware", "Sensor Y":"Electronics"}
base_prices = {"Widget A":14.99, "Widget B":22.50, "Gadget X":149.99,"Connector":4.99, "Sensor Y":89.99}

prod_col = np.random.choice(products, n)
df_raw   = pd.DataFrame({
    "date":       pd.date_range("2024-01-01", periods=n, freq="D"),
    "product":    prod_col,
    "category":   [categories[p] for p in prod_col],
    "quantity":   np.random.randint(5, 500, n),
    "unit_price": [round(base_prices[p] * np.random.uniform(0.95, 1.05), 2) for p in prod_col],
    "region":     np.random.choice(["North", "South", "East", "West"], n),
})
df_raw.to_csv("supply_chain.csv", index=False)
print("supply_chain.csv created.")
```

```python
# Arithmetic
df["revenue"]      = df["quantity"] * df["unit_price"]
df["revenue_k"]    = (df["revenue"] / 1000).round(2)   # in thousands

# Date extraction
df["month"]        = df["date"].dt.month
df["day_name"]     = df["date"].dt.day_name()
df["quarter"]      = df["date"].dt.quarter

# String operation
df["product_upper"] = df["product"].str.upper()
df["sku_code"]      = df["product"].str.replace(" ", "-")

# Apply a function to each value — like map() but for DataFrames
df["rev_band"] = df["revenue"].apply(
    lambda x: "high" if x > 5000 else "mid" if x > 1000 else "low"
)

print(df[["product", "revenue", "rev_band", "month"]].head(6))
```

```python
# .iloc — by integer position (0-based, like NumPy)
print(df.iloc[0])             # first row as a Series
print(df.iloc[0, 1])          # first row, second column
print(df.iloc[:5])             # first 5 rows (same as .head(5))
print(df.iloc[:5, :3])         # first 5 rows, first 3 columns
print(df.iloc[[0, 5, 10]])    # rows 0, 5, and 10

# .loc — by label (column name, index value)
print(df.loc[0, "product"])          # row 0, column "product"
print(df.loc[:4, ["product", "revenue"]])  # rows 0-4, two columns

# .at and .iat — fast single cell access
print(df.at[0, "product"])          # faster than .loc for a single cell
print(df.iat[0, 1])                 # faster than .iloc for a single cell
```

```python
df = pd.read_csv("supply_chain.csv", parse_dates=["date"])
df["revenue"] = df["quantity"] * df["unit_price"]

# Single condition
hardware   = df[df["category"] == "Hardware"]
high_rev   = df[df["revenue"] > 5000]
not_north  = df[df["region"] != "North"]
print(f"Hardware rows: {len(hardware)}")
print(f"High revenue rows: {len(high_rev)}")

# Multiple conditions — use & (and) | (or) ~ (not)
hw_north = df[(df["category"] == "Hardware") & (df["region"] == "North")]
big_orders = df[(df["quantity"] > 300) | (df["revenue"] > 10000)]
print(f"Hardware in North: {len(hw_north)}")

# .isin() — match against a list of values
widgets = df[df["product"].isin(["Widget A", "Widget B"])]
print(f"Widget rows: {len(widgets)}")

# .between() — range filter on a numeric column
mid_size = df[df["quantity"].between(100, 300)]

# String contains — partial match
gadgets  = df[df["product"].str.contains("Gadget")]

# Date filtering
q1 = df[df["date"] < "2024-04-01"]
print(f"Q1 rows: {len(q1)}")
```

```python
# Equivalent to df[(df["category"]=="Hardware") & (df["revenue"]>3000)]
result = df.query('category == "Hardware" and revenue > 3000')
print(result.head(3))

# Use @ to reference a Python variable inside the query string
min_qty = 200
high_qty = df.query("quantity >= @min_qty")

# List membership in query
target_regions = ["North", "East"]
ne_orders = df.query("region in @target_regions")
print(f"N+E orders: {len(ne_orders)}")
```
