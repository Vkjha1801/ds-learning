Exercise 2.1 — Sales list analyser

Create a list of 7 daily sales figures (make up realistic numbers between 500 and 5000). Using only list operations and built-in functions, compute and print:

— Total sales for the week
— Daily average
— Best day's sales and worst day's sales
— Sales on the first day and last day
— The top 3 days (sorted highest first)

```python
sales = [520,890,1890,501,4560,3200,2600]

print(f"Total Sales for the week: {sum(sales):,}")
print(f"Daily Average: {sum(sales)/len(sales):,}")
print(f"Best Day's Sale: {max(sales)}, Worst Day's sales: {min(sales):,}")
print(f"First day sale: {sales[0]:,}, Last day sales: {sales[6]:,}")

sort_sales = sorted(sales, reverse = True)

print(f"Top 3 days sales: {sort_sales[0]}, {sort_sales[1]},{sort_sales[2]}")
```

Exercise 2.2 — Product catalogue dict

Create a list of 4 product dictionaries. Each product must have: name, category, price (float), quantity (int), in_stock (bool). Then write code that:

— Prints every product's name and price in a formatted line
— Prints only the in-stock products
— Computes and prints the total inventory value (price × quantity) across all products

```python
Products = [
    {"name":"Apple","category":"Fruit","price": 2, "qty": 2000, "in_stock": True},
    {"name":"Laptop","category":"Electronics","price": 1000, "qty": 15, "in_stock": True},
    {"name":"Onion","category":"Vegetable","price": 1.5, "qty": 1500, "in_stock": False},
    {"name":"Tee","category":"Apparel","price": 30, "qty": 350, "in_stock": True}
    ]

print("Name and price of all products")
for p in Products:
    print(f"Name: {p["name"]:12} Price: {p["price"]}") 

print("\nList of products in stock")
for p in Products:
    if p["in_stock"]:
        print(f"Name: {p["name"]}")

print("\nTotal Inventory Value of each product")
for p in Products:
    Value = p["price"] * p["qty"]
    print(f"{p['name']} : ${Value:.2f}")
```

Exercise 2.3 — Duplicate finder with sets

You have two lists of order IDs — one from System A and one from System B. Using set operations, find:

— Orders in both systems (matched)
— Orders only in System A (missing from B)
— Orders only in System B (missing from A)
— All unique order IDs combined

Print a clear summary report showing each result and its count.


```python



SA = [1221,1222,1225,1228,1229]
SB = [1220,1222,1228]

SystemA = set(SA)
SystemB = set(SB)

C1 = SystemA & SystemB
C2 = SystemA - SystemB
C3 = SystemB -SystemA
C4 = SystemA | SystemB

print(f"Orders in both systems: {C1} Count: {len(C1)}")
print(f"Orders only in System A: {C2} Count: {len(C2)}")
print(f"Orders only in System B: {C3} Count: {len(C3)}")
print(f"All unique order IDs combined: {C4} Count: {len(C4)}")

```
