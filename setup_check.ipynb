```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(f"NumPy:      {np.__version__}")
print(f"Pandas:     {pd.__version__}")
print("Data libraries OK")
```

```python
df = pd.DataFrame({
    "product": ["Widget A", "Widget B", "Widget C"],
    "sales":   [120, 340, 89]
})
print(df)
print(f"\nBest seller: {df.loc[df.sales.idxmax(), 'product']}")
print("Pandas OK")
```

```python
plt.figure(figsize=(6, 3))
plt.bar(df["product"], df["sales"], color="#378ADD")
plt.title("Sales by product")
plt.tight_layout()
plt.show()
print("Matplotlib OK")
```
