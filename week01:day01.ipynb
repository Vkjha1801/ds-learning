Exercise 1.2 — Temperature converter

Write a script that asks the user to enter a temperature in Celsius using input(). Convert it to Fahrenheit (F = C × 9/5 + 32) and Kelvin (K = C + 273.15). Print all three values formatted to 2 decimal places with labels.


```python
temperature_C = float(input("Enter Temperature:"))
temperature_F = (temperature_C * 9/5) + 32
temperature_K = temperature_C + 273.15

print(f"Temperature in Celsius is {temperature_C:.2f} °C")
print(f"Temperature in Celsius is {temperature_F:.2f} °F")
print(f"Temperature in Celsius is {temperature_K:.2f} K")
```

Exercise 1.3 — Invoice calculator

Create variables: product_name, quantity (int), unit_price (float), discount_pct (float, e.g. 10 for 10%), tax_rate (float, e.g. 18 for 18%). Calculate and print a formatted invoice showing: subtotal, discount amount, price after discount, tax amount, and final total. Use f-strings with 2 decimal places throughout.

```python
product_name = input("Enter Product Name")
quantity = int(input("Enter the qty:"))
unit_price = float(input("Enter the price per unit:"))
discount_pct = float(input("Enter discount percentage:"))
tax_rate = float(input("Enter the tax rate in percentage:"))

subtotal = quantity * unit_price
disc_amt = subtotal * discount_pct/100
price_after_disc = subtotal - disc_amt
tax_amt = price_after_disc * tax_rate/100
final_total = price_after_disc + tax_amt

print(f"=== Invoice: {product_name} ===")
print(f"Qty x Price: {quantity} X {unit_price:.2f}")
print(f"Subtotal: {subtotal:.2f}")
print(f"Discount Amount: {disc_amt:.2f}")
print(f"Price After Discount: {price_after_disc:.2f}")
print(f"Tax Amount: {tax_amt:.2f}")
print(f"Final Total: {final_total:.2f}")
```
