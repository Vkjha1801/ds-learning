Exercise 4.1 — Metric calculator function library

Write a set of functions that form a mini analytics library. Each function should have a docstring, type hints, and handle edge cases gracefully:

— calculate_mean(numbers) → float
— calculate_median(numbers) → float
— calculate_std(numbers) → float (standard deviation)
— summarise(numbers, label="Dataset") → prints a full summary report

Test each function with a list of 7 daily sales figures.

```python
import math

def calculate_mean(numbers):
    """Return the arithmetic mean of a list of numbers"""
    if len(numbers) == 0:
        return 0
    else:
        return (sum(numbers)/len(numbers))

def calculate_median(numbers):
    """Return the median of a list of numbers"""
    sortnumbers = sorted(numbers)
    n = len(numbers)
    #print(n)

    if n%2 == 1:
        return sortnumbers[int((n+1)/2) - 1]
    else:
        return round((sortnumbers(n) + sortnumbers(n+1))/2,1)

def calculate_std(numbers):
    """Return the standard deviation of a list of numbers"""
    mean = calculate_mean(numbers)
    meanlist = [mean for x in numbers]
    variance = [((n-p)**2) for n,p in zip(numbers,meanlist)]

    return math.sqrt(sum(variance)/(len(variance)-1))

def summarise(numbers, label = "Dataset"):
    """Summarises the given list of numbers"""
    meanval = calculate_mean(numbers)
    medianval = calculate_median(numbers)
    stdval = calculate_std(numbers)

    print(f"{label} -> {numbers}")
    print(f"The mean values is {meanval}")
    print(f"The median value is {medianval}")
    print(f"The standard deviation is {stdval}")


summarise([6,19,36,71,54,2,99])

        
```

Exercise 4.2 — Refactor Day 2's product catalogue

Go back to Exercise 2.2 from Day 2 (the product catalogue with a list of dicts). Rewrite it so that each piece of logic lives in its own function:

— get_in_stock(products) → returns a filtered list
— total_inventory_value(products) → returns a float
— sort_by_price(products, ascending=True) → returns sorted list
— print_catalogue(products) → prints formatted output
— main() → calls all the above in order

Run the analysis by calling main() at the bottom. Compare this version to your Day 2 script — notice how much easier it is to read and modify.

```python
Products = [
    {"name":"Apple","category":"Fruit","price": 2, "qty": 2000, "in_stock": True},
    {"name":"Laptop","category":"Electronics","price": 1000, "qty": 15, "in_stock": True},
    {"name":"Onion","category":"Vegetable","price": 1.5, "qty": 1500, "in_stock": False},
    {"name":"Tee","category":"Apparel","price": 30, "qty": 350, "in_stock": True}
    ]

def get_in_stock(Product_Dict):
    """Return only in-stock products."""
    Instock = [p for p in Product_Dict if p["in_stock"]]
    return Instock

def total_inventory_value(Product_Dict):
    """Return total inventory value (price × quantity) across all products."""
    value = [p["price"] * p["qty"] for p in Product_Dict]
    return sum(value)

def sort_by_price(Product_Dict, ascending = True):
    """Return products sorted by price."""
    by_price = sorted(Product_Dict, key=lambda p: p["price"])
    return by_price

def print_catalogue(Product_Dict):
    """Print a formatted product catalogue."""
    print(f"{'Name':12} {'Price':>8} {'Qty':>6} {'Item in Stock'}")
    print("-" * 38)
    for p in Product_Dict:
        print(f" {p['name']:12}  ${p['price']:>7.2f} {p['qty']:>6} {p['in_stock']}")
        


def main():
    """Run the full product catalogue analysis."""
    print("=== Full catalogue (sorted by price) ===")
    print_catalogue(sort_by_price(Products))

    print("\n=== In-stock products only ===")
    print_catalogue(get_in_stock(Products))

    total = total_inventory_value(Products)
    print(f"\nTotal inventory value: £{total:,.2f}")





main()
```
